```python
import numpy as np
import pandas as pd
```

```python
# csv의 의미: comma-separated-values 
# - 참고 : https://ko.wikipedia.org/wiki/CSV_(%ED%8C%8C%EC%9D%BC_%ED%98%95%EC%8B%9D), https://ecommerce-platforms.com/ko/glossary/comma-separated-values-csv
# 확장자? : https://ko.wikipedia.org/wiki/%ED%8C%8C%EC%9D%BC_%ED%99%95%EC%9E%A5%EC%9E%90
# - 참고 : https://pandajeytv.tistory.com/113, https://mansoostat.tistory.com/63
df = pd.read_csv('법정동코드 전체자료.txt', sep='\t', encoding='euc-kr')
df.head()
```

```python
df.shape
```

```python
# df 다시 실행했더니 됐음!
df = df[df.폐지여부 == '존재']
df.shape
```

```python
# 인덱싱
df.법정동명.str[:2].head()
```

```python
# 찾기
# find(): 이 str안에 '동'이라는 글자가 있으면 12, 없으면 -1을 return
    # 결과값 >= 0 : 내가 찾는 문자열이 존재한다.
    # if문에서 find한 결과 >= 0 라는 형태로 자주 쓰인다.
df.법정동명.str.find('동').head()
```

```python
# 문자 대체 : replace()
df.법정동명.str.replace(' ',"_").head()
```

```python
# 예시
s = pd.Series(['가나다 123', 'ㅠㅠ!!!!!', '한글alphabet'])
s
```

```python
# 정규 표현식(Regular Expression) - 암호 등 지정할 때 문자열 패턴을 찾아내주는 것을 의미
    # 수업 자료 : https://wikidocs.net/book/2155 에서 02-05) 정규표현식 참고
import re
re.sub('[0-9]', '.', '가나다123')    # re.sub(정규표현식, 대체할 문자, 대상)
```

```python
re.sub('[a-z]', '.', '한글alphabet')
```

```python
# 문자열 메소드인 replace는 정규표현식을 지원하지 않는다.
'한글alphabet'.replace('[a-z]', '.')
```

```python
# 한글 정규표현식
    # 그 밖에도 이메일 정규표현식이 있다. ex. email 정규표현식 파이썬 으로 검색
re.sub('[ㄱ-ㅎㅏ-ㅣ가-힣]', '.', '한글alphabet')
```

```python
# Pandas str.replace() 메소드는 정규표현식을 지원함
s.str.replace('[ㄱ-ㅎㅏ-ㅣ가-힣]', '.') 
```

```python
# futurewarning이 떠서 적용이 안됨! - 이유 찾아보기!!
# ^는 not(부정)을 의미함
    # \s == white space를 의미
    #    != 공백
s.str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣 ] ', '') # 한글과 공백을 제외한 문자열은 삭제한다.
```

- white space = ' \t\n'

- str.메소드로 천단위 구분기호를 숫자로 바꾸기

```python
cctv = pd.read_csv('서울시CCTV설치운영현황(자치구)_년도별_210731기준.csv',
                    skiprows=1, encoding='euc-kr')
cctv.head()
```

```python
for column in cctv.columns[1:]:
    cctv[column] = cctv[column].str.replace(',','').astype(int)
cctv.head()
```

```python
cctv.index[:5]
```

- 인덱스 조작

```python
cctv.set_index('구분').head()
```

```python
# set_index()는 자기파괴적 메소드이므로, 데이터 원본 자체는 변하지 않는다.
cctv.head()
```

```python
cctv.set_index('구분', inplace=True)
cctv
```

```python
# set_index로 수정한 부분을 원래대로 바꾸려면
cctv.reset_index('구분', inplace=True)
cctv
```

- set_index를 한 후에 주의사항

```python
cctv.set_index('구분', inplace=True)
cctv.head()
```

```python
cctv.to_csv('cctv.csv')# cctv라는 excel 파일이 생성되어 있다
cctv2 = pd.read_csv('cctv.csv')
cctv2.head()
```

```python
# Unnamed ? - 실행하고 csv 파일 확인해보기
cctv2.to_csv('cctv.csv') 
cctv3 = pd.read_csv('cctv.csv')
cctv3.head()
```

```python
# Unnamed:0 인덱스를 없애려면, index=None을 설정
cctv2.to_csv('cctv.csv', index=None)
cctv3 = pd.read_csv('cctv.csv')
cctv3.head()
```
