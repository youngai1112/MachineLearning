```python
ans.shape # 행이 44개, 열이 3개인 DataFrame
```

- dataset 1 에 대해서 그래프를 그려보자.

```python
ds = ans[ans.dataset == 'I']  # 원하는 조건을 filtering해서 DF로 만들기
ds
```

- 회귀식 y = 0.5 * x + 3

```python
plt.scatter(ds.x, ds.y)                 # 산점도
plt.plot([4,14], [5,10], 'r-', lw=2)    # 실선 좌표 설정
plt.title('Dataset I')
plt.show()
```

- 4개의 그래프를 하나의 그림에 그리기


```python
# subplot()과 subplots()의 차이점
#   - https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=heygun&logNo=221520454294
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ax = axs[i//2, i%2]
    ax.plot([4,14],[5,10], 'r-', lw=2) # lw: line width
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
for i, roman in enumerate(['I', 'II', 'III', 'IV']):    # enumerate는 index값을 받기 때문에 i를 변수로 둔다.
    ax = axs[i//2, i%2]
    ds = ans[ans.dataset == roman]
    ax.scatter(ds.x, ds.y)
    ax.plot([4,14],[5,10], 'r-', lw=2) 
    ax.set_title(f'Dataset {roman}')
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
for i, roman in enumerate(['I', 'II', 'III', 'IV']):    # enumerate는 index값을 받기 때문에 i를 변수로 둔다.
    ax = axs[i//2, i%2]
    ds = ans[ans.dataset == roman]
    ax.scatter(ds.x, ds.y)
    ax.plot([4,20],[5,13], 'r-', lw=2) # 스케일이 달라서 숫자를 수정하였다.
    ax.set_title(f'Dataset {roman}')
```

```python
# 4개의 그래프가 scale이 동일해졌다. 
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
for i, roman in enumerate(['I', 'II', 'III', 'IV']):    # enumerate는 index값을 받기 때문에 i를 변수로 둔다.
    ax = axs[i//2, i%2]
    ds = ans[ans.dataset == roman]
    ax.scatter(ds.x, ds.y)
    ax.plot([3,20],[4.5,13], 'r-', lw=2) 
    ax.set_ylim(2,14)
    ax.set_title(f'Dataset {roman}')
```
