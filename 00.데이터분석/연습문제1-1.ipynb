```python
# b. 
# pandas로 4분위 수 및 4분위 범위 구하기 참고 : https://sosomemo.tistory.com/34
# 이상치 제거하기 참고 : https://hungryap.tistory.com/69, https://fhaktj8-18.tistory.com/entry/%ED%8C%90%EB%8B%A4%EC%8A%A4-pandas-IQR-%ED%99%9C%EC%9A%A9%ED%95%B4%EC%84%9C-%EC%9D%B4%EC%83%81%EC%A0%90outlier-%EC%B0%BE%EA%B3%A0-%EC%82%AD%EC%A0%9C%ED%95%98%EA%B8%B0,
#                       https://fhaktj8-18.tistory.com/entry/%ED%8C%90%EB%8B%A4%EC%8A%A4-pandas-IQR-%ED%99%9C%EC%9A%A9%ED%95%B4%EC%84%9C-%EC%9D%B4%EC%83%81%EC%A0%90outlier-%EC%B0%BE%EA%B3%A0-%EC%82%AD%EC%A0%9C%ED%95%98%EA%B8%B0

# q1은 Setosa, Versicolour, Virginica의 25% 범위
```

```python
# setosa - q1, q3, iqr, upper_bound, lower_bound 구하기
setosa = iris[iris.species == 'setosa']
setosa.head(3)
```

```python
setosa1 = setosa.iloc[:, :-1]
setosa1.head()
```

```python
setosa_q1 = setosa.quantile(0.25)
setosa_q1
```

```python
setosa_q3 = setosa.quantile(0.75)
setosa_q3
```

```python
setosa_iqr = setosa_q3 - setosa_q1
setosa_iqr
```

```python
setosa_lower_bound = setosa_q1 - (setosa_iqr * 1.5)
setosa_lower_bound
```

```python
setosa_lb_outlier1 = setosa1[setosa1 < setosa_lower_bound]
setosa_lb_outlier1
```

```python
setosa_lb_outlier2 = setosa_lb_outlier1.dropna(how='all').index
setosa_lb_outlier2
```

```python
setosa_upper_bound = setosa_q3 + (setosa_iqr * 1.5)
setosa_upper_bound
```

```python
setosa_ub_outlier1 = setosa1[setosa1 > setosa_upper_bound]
setosa_ub_outlier1
```

```python
setosa_ub_outlier2 = setosa_ub_outlier1.dropna(how='all').index
setosa_ub_outlier2
```

```python
setosa1.drop(setosa_lb_outlier2, inplace=True)
```

```python
setosa1.drop(setosa_ub_outlier2, inplace=True)
```

```python
setosa1
```

# 2. Titanic

```python
# a. 
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
titanic['category 1'] = np.where(titanic['age'].values >= 20, titanic.sex, 'child')
titanic
```

```python
# b. 
titanic.age.value_counts
```

```python
titanic_age_mean = titanic.age.mean()
titanic_age_mean
```

```python
titanic_age = titanic.age.fillna(round(titanic_age_mean))
titanic_age.isna().sum()
```

```python
# c. 
titanic_groupby = titanic.groupby(['sex', 'class', 'embark_town'])[['survived']].mean()
titanic_groupby
```

# 3. Mile Per Gallon

```python
# a. 
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
mpg['hp_per_cc'] = mpg.horsepower / mpg.displacement
mpg
```

```python
# b. 
mpg.name.str.split()[0][0]
```

```python
mpg1 = mpg.name.str.split()[0][1:]
mpg1 = ' '.join(mpg1)
mpg1
```

```python
mpg['manufacturer'] = mpg.name.apply(lambda x : x.split()[0])
mpg
```

```python
mpg['model'] = mpg.name.apply(lambda x: ' '.join(x.split()[1:]))
mpg
```

```python
mpg.drop(columns=['name'])
```

```python
# c. 
mpg_cylinders_mean = mpg.groupby(['cylinders'])['mpg'].mean()
mpg_cylinders_mean = pd.DataFrame(mpg_cylinders_mean)
mpg_cylinders_mean
```

```python
# d. 
mpg_origin_groupby = mpg.groupby(['origin'])[['hp_per_cc']].mean()
mpg_origin_groupby
```

```python
# e. value_counts() > 5 --> boolean type을 DF에 넣어서 조건에 맞게 정렬하기
mpg_vc = pd.DataFrame(mpg.manufacturer.value_counts() > 5)
mpg_vc
```

```python
mpg_vc_5 = pd.DataFrame(mpg_vc[mpg_vc])
mpg_vc_5
```

```python
mpg_mpg_mean = mpg.groupby(['manufacturer']).mean()
mpg_mpg_mean
```
