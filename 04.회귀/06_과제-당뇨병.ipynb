# 과제 - 당뇨병 데이터로 다중 회귀

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head(3)
# feature가 10개, target이 연속적인 값으로 나타나 있는 회귀에 적합한 데이터
```

- 각 feature별 산점도와 선형회귀선

```python
_, axs = plt.subplots(nrows=2, ncols=5, figsize=(16,8))
for i, feature in enumerate(df.columns[:-1]):
    row, col = i//5, i%5   # col의 개수 == 5개
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- 각 feature 별 R squared value 구하기

```python
from sklearn.linear_model import LinearRegression
```

```python
# 개별적인 도출형식 확인하고 for문 돌리기!
y = diabetes.target
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1,1) # df[feature]는 1차원 데이터 -> Series를 2차원으로 변경 불가능! -> .values.reshape(-1,1) 행이 여러개니까 -1, 열은 1개
    lr = LinearRegression()
    lr.fit(X,y)
    r2 = lr.score(X,y)
    print(f'{feature}: {r2:.4f}')
```

- dataset 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    diabetes.data, diabetes.target, test_size=0.1, random_state=2021
)

X_train.shape ,X_test.shape, y_train.shape, y_test.shape
```

### Regressor 별로 MSE 비교

```python
from sklearn.linear_model import LinearRegression       # 선형회귀
from sklearn.svm import SVR                             # Support Vector Regressor
from sklearn.ensemble import RandomForestRegressor      # RandomForestRegressor
from sklearn.metrics import mean_squared_error          # MSE
```

```python
# Linear Regressor
lr = LinearRegression()
lr.fit(X_train ,y_train)
pred_lr = lr.predict(X_test)
mse_lr = mean_squared_error(y_test, pred_lr)    # 실제값과 예상값의 차이 구하기, 낮을수록 좋은 모델
```

```python
# SVR
svr = SVR()
svr.fit(X_train ,y_train)
pred_sv = svr.predict(X_test)
mse_sv = mean_squared_error(y_test, pred_sv)    # 실제값과 예상값의 차이 구하기, 낮을수록 좋은 모델
```

```python
# Random Forest Regressor
rfr = RandomForestRegressor()
rfr.fit(X_train ,y_train)
pred_rf = rfr.predict(X_test)
mse_rf = mean_squared_error(y_test, pred_rf)    # 실제값과 예상값의 차이 구하기, 낮을수록 좋은 모델
```

- 결과 비교

```python
# MSE
print(f'LR: {mse_lr:.4f}, SVR: {mse_sv:.4f}, RFR: {mse_rf:.4f}')
```

```python
# 실제값과 예측값 비교
df = pd.DataFrame({
    'y_test':y_test, 'LR':pred_lr, 'SVR':pred_sv, 'RFR':pred_rf
})
df.head(10)
```
